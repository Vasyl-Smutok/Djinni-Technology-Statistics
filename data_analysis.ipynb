```python
import pandas as pd
from parse import get_requirements_for_all_jobs
```

```python
def get_technologies_statistics(url: str, plot_bar_color: str) -> None:
    data = get_requirements_for_all_jobs(url)
    df = pd.DataFrame({"NAME": data})

    df["TECHNOLOGY'S"] = df.index
    df["NAME"] = df["NAME"].str.upper()

    df = df.groupby(by="NAME").count().sort_values(by="TECHNOLOGY'S", ascending=False).head(25)

    df.plot.bar(
        figsize=(15 ,7),
        color=plot_bar_color,
        title=f"{url.split('=')[-1].upper()} TECHNOLOGY'S"
    )
```

```python
get_technologies_statistics("https://djinni.co/jobs/?primary_keyword=Python", "gold")
```

```python
get_technologies_statistics("https://djinni.co/jobs/?primary_keyword=Java", "blue")
```

```python
get_technologies_statistics("https://djinni.co/jobs/?primary_keyword=JavaScript", "green")
```
